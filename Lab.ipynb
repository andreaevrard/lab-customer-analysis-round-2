5. Check and deal with NaN values.

```python
df.isna().sum()
```

```python
round(df.isna().sum()/len(df),4)*100
```

```python
nulls_df = pd.DataFrame(round(df.isna().sum()/len(df),4)*100)
nulls_df
nulls_df = nulls_df.reset_index()
nulls_df
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
columns_drop = nulls_df[nulls_df['percent_nulls']>6]['header_name']  
print(columns_drop.values)
```

```python
df = df[df['state'].isna()==False] 
df = df[df['response'].isna()==False]
df = df[df['months since last claim'].isna()==False] 
df = df[df['number of open complaints'].isna()==False] 
df = df[df['vehicle size'].isna()==False] 
df = df[df['vehicle class'].isna()==False] 
df
```

```python
df = df.drop(['vehicle type'], axis=1)
```
